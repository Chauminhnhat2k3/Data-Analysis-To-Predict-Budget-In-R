```r
#Simple Linear Regression
```

```r
# Define train dataset with > 40000 as the Budget
train_data <- train_data %>%
  filter(Time_taken != "NA", Budget >30000 )

head(train_data)
```

```r
linear_model_train <- lm(Time_taken ~ Budget, data = train_data)
```

```r
summary(linear_model_train)
```

```r
# Input data we use to predict budget in train data
new_budget_train <- data.frame(
  Budget = c(2000, 3000, 4000))

# Predict the data points
pred_train <- predict(linear_model_train, newdata = new_budget_train, interval = "confidence")
pred_train
```

```r
# Define test dataset with > 30000 as the Budget
test_data <- test_data %>%
  filter(Time_taken != "NA", Budget >30000 )

head(test_data)
```

```r
linear_model_test <- lm(Time_taken ~ Budget, data = test_data)
```

```r
summary(linear_model_test)
```

```r
# Input data we use to predict budget in test data
new_budget_test <- data.frame(
  Budget = c(2000, 3000, 4000))

# Predict the test data points
pred_test <- predict(linear_model_test, newdata = new_budget_test, interval = "confidence")
pred_test
```

```r
#1. Model Evaluation
#1.1 Training and Testing Data
```

```r
library(rlang)
```

```r
library(tidymodels)
```

```r
# Library for modeling
library(tidymodels)

# Load tidyverse
library(tidyverse)
```

```r
# Pick linear regression
lm_spec <- linear_reg() %>%
  # Set engine
  set_engine(engine = "lm")

# Print the linear function
lm_spec
```

```r
train_fit <- lm_spec %>% 
    fit(Time_taken ~ Budget, data = train_data)

train_fit 
```

```r
train_results <- train_fit %>%
  # Make the predictions and save the predicted values
  predict(new_data = train_data) %>%
  # Create a new column to save the true values
  mutate(truth = train_data$Time_taken)

head(train_results)
```

```r
test_results <- train_fit %>%
  # Make the predictions and save the predicted values
  predict(new_data = test_data) %>%
  # Create a new column to save the true values
  mutate(truth = test_data$Time_taken)

head(test_results)
```

```r
#1.3 Evaluating the Model
```

```r
rmse(train_results, truth = truth,
     estimate = .pred)
```

```r
rmse(test_results, truth = truth,
     estimate = .pred)
```

```r
rsq(train_results, truth = truth,
    estimate = .pred)
```

```r
rsq(test_results, truth = truth,
    estimate = .pred)
```

```r
test_results %>%
  mutate(train = "testing") %>%
  bind_rows(train_results %>% mutate(train = "training")) %>%
  ggplot(aes(truth, .pred)) +
  geom_abline(lty = 2, color = "orange", 
              size = 1.5) +
  geom_point(color = '#006EA1', 
             alpha = 0.5) +
  facet_wrap(~train) +
  labs(x = "Truth", 
       y = "Predicted Budget (min)")
```
